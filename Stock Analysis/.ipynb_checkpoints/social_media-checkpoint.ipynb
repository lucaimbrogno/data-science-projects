This is a project that I wanted to do to get a little more familiar with pandas, matplotlib, and seaborn libraries in Python. I picked a few stocks that I am interested in and did some analysis on these stocks. A few things I wanted to cover are:

```python
# General necessary libraries
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style('whitegrid')
%matplotlib inline

# With DataReader we can pull a whole bunch of stock information from yahoo
from pandas_datareader import DataReader


from datetime import datetime
```

# Stock Selection
Here we will choose which stocks we want to analyze. Since I recently completed an web development internship at a company in the social media influencer space, I think it would be interesting to look at a few popular social media companies! 
- Facebook
- Snapchat
- Twitter

```python
stocks = ['FB', 'SNAP', 'TWTR']

# retrieve last year of data for each stock from google finance
end_date = datetime.now()
start_date = datetime(end_date.year-2, end_date.month, end_date.month)

stock_data = {}
for stock in stocks:
    stock_data[stock] = DataReader(stock,'yahoo',start_date,end_date)
```

```python
stock_data['FB'].describe()
```

# Risk Analysis
In investing, risk is commonly defined as the measure of any deviation from an expected outcome. There are many ways to measure the risk of an investment. Two metrics I'm going to focus on in this analysis are:
- Standard deviation: The most obvious metric given the definition above. Measures the dispersion around the mean.
- Beta: A measure of a stocks volatility in relation to the overall market. The market. in theory, has a beta of 1.0


The previous figure displayed above shows the daily return of Facebook stock over time. Continuing in that direction, we can use our stocks array to calculate the daily percentage change for all three stocks and then calculate the standard deviation from there.

```python
stocks_close = DataReader(stocks,'yahoo',start_date,end_date)['Adj Close']
stocks_returns = stocks_close.pct_change()

stocks_returns.std()
```

From this metric, Facebook has the lowest risk level. However it may be useful to use more than two years data to compare the risk of these stocks. Another way to look at this type of risk is with a histogram. With a histogram we can see our confidence level for achieving different levels of returns.

```python
sns.distplot(stocks_returns['FB'].dropna(),bins=80)
```

```python
# The 0.05 quantile of daily returns for FB
stocks_returns['FB'].quantile(0.05)
```

So this value means that with 95% confidence, our daily loss will not exceed 3.3%. To compare, we can calculate the same metric for SNAP and TWTR as well.

```python
# The 0.05 quantile of daily returns for SNAP
stocks_returns['SNAP'].quantile(0.05)
```

```python
# The 0.05 quantile of daily returns for TWTR
stocks_returns['TWTR'].quantile(0.05)
```

Now let's take a look at Beta to observe another measure of risk. The way we will calculate the Beta for our stocks is to plot a regression line between our stock's returns and the market returns. The beta will be the slope of the regression line between our stocks returns and the markets returns. 

First we have to combine the daily returns of these three stocks as well as the market return into a single data structure. We will use the SPY index to represent the market.

```python
# First lets add the S&P 500 to our list of stocks
stocks.append('SPY')
```

```python
# Combine daily closing prices of each stock into one data frame
stocks_close = DataReader(stocks,'yahoo',start_date,end_date)['Adj Close']

stocks_close.head()
```

```python
# Convert to daily percentage gain/loss
stocks_returns = stocks_close.pct_change()

# Let's take a look at our data
stocks_returns.head()
```

Using Seaborn, we can create a scatter plot comparing the daily returns of FB vs SPY. We can also generate a regression line in the data.

```python
fb_spy_plot = sns.jointplot('SPY','FB',stocks_returns, kind='regplot',height=8)
```

Finally, let's check out the slope of the regression line to get our Beta! We will have to use a different library since seaborn is mainly for data visualizations.

```python
from scipy.stats import linregress

x = stocks_returns['SPY'].dropna().values
y = stocks_returns['FB'].dropna().values

# Getting slope of regression line between SPY and FB returns
fb_beta = linregress(x,y).slope
print('FB Beta: {}'.format(fb_beta))
```

```python
# Same for SNAP
y = stocks_returns['SNAP'].dropna().values
snap_beta = linregress(x,y).slope
print('SNAP Beta: {}'.format(snap_beta))
```

```python
# Same for TWTR
y = stocks_returns['TWTR'].dropna().values
twtr_beta = linregress(x,y).slope
print('TWTR Beta: {}'.format(twtr_beta))
```

As we can see, FB has the lowest Beta representignt he lowest risk among the three stocks. This is consistent with our results of risk measures using the standard deviation method as well.

# Comparison Analysis
Let's do some comparisons between these stocks. I like to use scatter plot matrices because it makes it really easy to see trends visually. Initially, we will look at the daily returns of each stock and how they compare with eachother.

```python
# Let's take a look at the dataframe we created above
stocks_returns.head()

# Remove SPY
stocks_returns = stocks_returns.drop('SPY',1)

stocks_returns.head()
```

```python
# Now let's look at the scatter plot of daily returns between facebook adn snapchat
sns.jointplot('FB','SNAP',stocks_returns, kind='scatter',height=8)
```

```python
# seaborn and Pandas allow us to build a full scatter plot matrix super easily!
sns.pairplot(stocks_returns.dropna(),height=3)
```

The histogram on the diagonal represents the distribution of daily percentage return for each stock. Observing the figure above, it looks like there exists a positive correlation between all three stocks and the highest correlation exists between Facebook and Twitter. To further investigate this we can try a slightly different visual representation.By using Seaborn's heatmap function we can see a numercial value representing the strength of the correlation.

```python
sns.heatmap(stocks_returns.corr(),annot=True,fmt=".3g")
```
